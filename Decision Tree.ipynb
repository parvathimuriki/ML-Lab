```python
#Impute the Age column. It has too many missing values but it may be important.
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import pandas as pd
import numpy as np

# Load the dataset
train_df = pd.read_csv("train.csv")

# Calculate mean and standard deviation of existing Age values
age_mean = train_df["Age"].mean()
age_std = train_df["Age"].std()

# Count missing values in Age
num_missing = train_df["Age"].isnull().sum()

# Generate random values from normal distribution
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=num_missing)

# Fill missing values with these random values
train_df.loc[train_df["Age"].isnull(), "Age"] = random_ages

# Verify if missing values are replaced
print(train_df["Age"].isnull().sum())  # Should print 0
```

```python
#Prepare X and y
import pandas as pd

# Load the dataset
train_df = pd.read_csv("train.csv")
y = train_df["Survived"]

# Step 2: Select Features (drop irrelevant columns)
X = train_df.drop(columns=["Survived", "PassengerId", "Name", "Ticket", "Cabin"], errors="ignore")

# Step 3: Encode Categorical Features
X = pd.get_dummies(X, columns=["Sex", "Embarked"], drop_first=True)  # One-hot encoding

# Display first few rows of X and y
print("Features (X):")
print(X.head())
print("\nTarget (y):")
print(y.head())
```

```python
# Label encoding of categorical features
import pandas as pd
from sklearn.preprocessing import LabelEncoder



# List of categorical features to encode
categorical_cols = ["Sex", "Embarked", "Pclass"]  # Add more if needed

# Fill missing values with a placeholder
train_df[categorical_cols] = train_df[categorical_cols].fillna("Unknown")

# Initialize the LabelEncoder
label_encoders = {}

# Apply Label Encoding to each categorical column
for col in categorical_cols:
    label_encoders[col] = LabelEncoder()
    train_df[col] = label_encoders[col].fit_transform(train_df[col])

# Display first few rows of the transformed dataset
print(train_df.head())
```

```python
# Create the model
# Fitting the decision tree with default hyperparameters
# Importing decision tree classifier from sklearn library


import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

# Load dataset
df = pd.read_csv("train.csv")
df = df.drop(columns=["Name", "Ticket", "Cabin", "PassengerId"]) 
df = pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first=True)
df = df.dropna() 

# Separate features (X) and target variable (y)
X = df.drop(columns=["Survived"])  # Replace "Target" with actual target column name
y = df["Survived"]  # Replace "Target" with actual target column name

# Split data: 90% Training, 10% Validation
X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

# Initialize Decision Tree Classifier with default hyperparameters
model = DecisionTreeClassifier(random_state=42)

# Train the model
model.fit(X_train, y_train)

# Predict on validation set
y_pred = model.predict(X_validation)

# Compute accuracy
accuracy = accuracy_score(y_validation, y_pred)

# Compute Confusion Matrix
cm = confusion_matrix(y_validation, y_pred)

# Print results
print("Decision Tree Classifier Performance:")
print("Accuracy:", accuracy)
print("Confusion Matrix:\n", cm)
```
